#  AULA 10 : MESCLANDO DATAFRAME

```python
import pandas as pd
```

```python
vendasJaneiro_DataFrame = pd.read_excel("Vendas_Jan.xlsx")
vendasFevereiro_DataFrame = pd.read_excel("Vendas_Fev.xlsx")
vendasMarco_DataFrame = pd.read_excel("Vendas_Mar.xlsx")


#Método append/concat -> permite inserir/unir DataFrames em outros DataFrames 
vendas_Jan_Fev = vendasJaneiro_DataFrame.append(vendasFevereiro_DataFrame)
display(vendas_Jan_Fev)
```

```python
vendas_geral_DF = pd.concat([vendasJaneiro_DataFrame,vendasFevereiro_DataFrame,vendasMarco_DataFrame]) 
display(vendas_geral_DF)
```

```python
#Keys -> permite dar nome os DataFrames concatenados
vendas_geral_DF_2 = pd.concat([vendasJaneiro_DataFrame,vendasFevereiro_DataFrame,vendasMarco_DataFrame], keys=['Janeiro','Fevereiro','Março']) 
display(vendas_geral_DF_2)
```

```python
#Imprimindo um mês em específico 
Fevereiro = vendas_geral_DF_2.loc['Fevereiro']
display(Fevereiro)
```
