# AULA 18: PIVOT TABLE

```python
import pandas as pd
```

```python
base_Lanchonete_DF = pd.read_excel("Vendas_Lanchonete_Pivot_Table.xlsx")
display(base_Lanchonete_DF)
```

```python
#Criação de uma tabela dinâmica com Pandas e Python


#index= Linhas
#Columns = Colunas
#Values = Soma
#Aggfunc = Tipo de calculo (sum = soma)
pivotExemplo1 = base_Lanchonete_DF.pivot_table(index="Data Venda", 
                                               columns="Cliente", values="Preço com Desconto", aggfunc="sum")
display(pivotExemplo1)
```

```python
pivotExemplo2 = base_Lanchonete_DF.pivot_table(index="Cliente", 
                                               columns="Data Venda", values="Preço com Desconto", aggfunc="sum")
display(pivotExemplo2)
```

```python
pivotExemplo3 = base_Lanchonete_DF.pivot_table(index="Cliente", 
                                               columns="Data Venda", values=["Preço Total", "Preço com Desconto"], aggfunc="sum")
display(pivotExemplo3)
```

```python
pivotExemplo4 = base_Lanchonete_DF.pivot_table(index="Data Venda", 
                                               columns=["Cliente", "Produto"], values="Preço com Desconto", aggfunc="sum")
display(pivotExemplo4)
```

```python
pivotExemplo5 = base_Lanchonete_DF.pivot_table(index="Data Venda", 
                    columns=["Cliente", "Produto"], values=["Preço Total","Preço com Desconto"], aggfunc="sum")

#Método fillna() -> me permite substituir os valores NaN(vazios) de um DataFrame por outro valor

pivotExemplo5["Preço com Desconto"] = pivotExemplo5['Preço com Desconto'].fillna(0) 
pivotExemplo5["Preço Total"] = pivotExemplo5['Preço Total'].fillna(0) 
display(pivotExemplo5)
```
