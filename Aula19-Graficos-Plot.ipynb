```python
import pandas as pd
import matplotlib.pyplot as grafico


#Data Frame Frutas
frutas_DF = pd.read_excel("Base_Grafico.xlsx")
display(frutas_DF)

```

```python
#Criando um gráfico

#plot = gráfico de linhas
grafico.plot(frutas_DF['Frutas'], frutas_DF['Total Vendas'])
grafico.title("Vendas Frutas") #Título do gráfico
grafico.xlabel("Nome das frutas") #Descrição do eixo x
grafico.ylabel("Quantidade de frutas") #Descrição do eixo y

#show = exibe o gráfico criado 

grafico.show()
```
