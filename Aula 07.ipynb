```python
import pandas as pd
import numpy as np
```

```python
Data_Frame = pd.read_excel("Tratamento_Dados.xlsx")

#Método fillna -> preenche os valores vazios de um DataFrame
#Método mean() -> retorna a média de determindos dados de um DataFrame

dadosData_Frame = Data_Frame.fillna(Data_Frame["Total Vendas"].mean())
display(dadosData_Frame)


#Método ffill -> preenche com o último resgistro válido encontrado em uma linha 
#ou coluna definida 

dados2Data_Frame = Data_Frame["Total Vendas"].ffill()
print(dados2Data_Frame)
```

```python
#Método value_counts -> realiza a contagem de valores dentro de um DataFrame

qtdVendas = Data_Frame['Vendedor'].value_counts()
print(qtdVendas)
```

```python
#Método groupby() -> agrupa informações 
#Método sum() -> realiza a soma de dados numéricos

vendaVendedor = Data_Frame.groupby("Vendedor").sum()
print(vendaVendedor)
```

```python
#OUTRO EXEMPLO
df = pd.DataFrame({'Animal': ['Falcon', 'Falcon',
                              'Parrot', 'Parrot'],
                   'Max Speed': [380., 370., 24., 26.]})
print(df.groupby(['Animal']).mean())
```

```python
#OUTRO EXEMPLO 2
arrays = [['Falcon', 'Falcon', 'Parrot', 'Parrot'],
          ['Captive', 'Wild', 'Captive', 'Wild']]
index = pd.MultiIndex.from_arrays(arrays, names=('Animal', 'Type'))
df = pd.DataFrame({'Max Speed': [390., 350., 30., 20.]},
                  index=index)
print(df.groupby(level=0).mean())

print()

df.groupby(level="Type").mean()
```
