# AULA 15: Merge Outer

```python
import pandas as pd
```

```python
vendasLoja_1_DF = pd.read_excel("Outer_Vendas_Loja1.xlsx")
vendasLoja_2_DF = pd.read_excel("Outer_Vendas_Loja2.xlsx")

#how = como
#on = coluna 
verificandoVendas_DF = pd.merge(vendasLoja_1_DF,vendasLoja_2_DF, on=["Id Vendedor"], how="outer", suffixes=('Loja 1', 'Loja 2'))
display(verificandoVendas_DF)
```

```python
#Removendo Linhas Vazias 
#MÃ©todo dropna -> Deleta as linhas que tem pelo menos 1 valor vazio

verificandoVendas_DF = verificandoVendas_DF.dropna()
display(verificandoVendas_DF)
```
