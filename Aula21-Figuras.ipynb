```python
import pandas as pd
import matplotlib.pyplot as grafico
```

```python
frutas_DF = pd.read_excel("Base_Grafico.xlsx")
```

```python
#Tamanho da figura que podemos aumentar ou diminuir

figura = grafico.figure(figsize=(50,10))

frutas = frutas_DF['Frutas']
total = frutas_DF['Total Vendas']


#Primeiro termo -> número de linhas
#Segundo termo -> número de colunas 
#Terceiro termo -> posição do gráfico
#label -> legenda 
figura.add_subplot(131)
grafico.plot(frutas, total, label='plot')
grafico.legend()
grafico.title("Gráfico 1")

for i in range(len(frutas)):
    grafico.annotate(frutas[i], (frutas[i], total[i]))
    
grafico.xticks([])
grafico.show()
```

```python
figura.add_subplot(131)
grafico.bar(frutas, total, label='plot')
grafico.legend()
grafico.title("Gráfico 2")

for i in range(len(frutas)):
    grafico.annotate(frutas[i], (frutas[i], total[i]))
    
grafico.xticks([])
grafico.show()
```

```python
grafico.pie(total, labels=frutas)
grafico.title("Gráfico 3")
```
