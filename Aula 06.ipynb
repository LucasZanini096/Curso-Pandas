# Aula 06 : Deletando Linhas e Colunas

```python
import pandas as pd
import numpy as np
```

```python
Data_Frame = pd.read_excel("Deletar_Linhas_Colunas.xlsx")
display(Data_Frame)
```

```python
#Posso fazer a remoção de valores vazios (NaN) de um DataFrame. O método
# .dropna() realiza a remoção de uma linha caso ela possui pelo menos 1 
#valor vazio. 
```

```python
Deletando_valores_vazios = Data_Frame.dropna(axis=0, how="all")
#0 -> simboliza as linhas (eixo x) #Vai retirar as linhas 3 e 8
#1 -> simboliza as colunas (eixo y) #Vai retirar a coluna Produto 
#how
#any -> retira se há ocorrência de pelo pelo menos 1 caso de NaN
#all -> retira se há a presença em todas as células de uma linha ou coluna 
#tiver NaN
display(Deletando_valores_vazios)
```

```python
#Posso retirar apenas uma coluna em específico 
del Data_Frame["Produto"]
print(Data_Frame)
```

```python
#Método Drop -> Método que me permite remover linhas ou colunas de um DataFrame
#Excluindo colunas
deletarDuasColunas = Deletando_valores_vazios.drop(columns=['Data Venda', 'Produto'])
display(deletarDuasColunas)
```

```python
#EXEMPLO
df = pd.DataFrame(np.arange(12).reshape(3, 4),
                  columns=['A', 'B', 'C', 'D'])
print(df.drop(['B', 'C'], axis=1))
```

```python
#Excluindo linhas em DataFrames
excluirLinha3 = Data_Frame.drop(2,axis=0) #Devo informar o eixo e a label ou index da linha ou coluna
print(excluirLinha3)
print()

excluir_Mais_Linhas = Data_Frame.drop([0,1])
print(excluir_Mais_Linhas)
```
