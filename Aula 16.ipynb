# AULA 16 : GROUPBY

```python
import pandas as pd
```

```python
vendas_DF = pd.read_excel("Groupby.xlsx")

#Groupby - Agrupa pela coluna de Vendedor e usando mean para calcular a média 
mediaVendedor = vendas_DF.groupby(['Vendedor']).mean()
display(mediaVendedor)
```

```python
#Groupby - Agrupa pela coluna de Vendedor e usando mean para calcular a soma 
somaVendedor = vendas_DF.groupby(['Vendedor']).sum()
display(somaVendedor)
```

```python
#groupby = Agrupa pela coluna
#dropna = deleta as linhas que tem pelo menos um valor vazio
#by = usar a coluna como critério para fazer o groupby
#sum = soma
deixandoValoresemBranco = vendas_DF.groupby(by=['Vendedor'], dropna=False).sum()
display(deixandoValoresemBranco)
```

```python
removendoValoresemBranco = vendas_DF.groupby(by=['Vendedor'], dropna=True).sum()
display(removendoValoresemBranco)
```

```python
agrupaDuasColunas = vendas_DF.groupby(["Vendedor","Produto"]).sum()
display(agrupaDuasColunas)
```

```python
agrupaFrutasVendedor = vendas_DF.groupby(['Produto','Vendedor']).sum()
display(agrupaFrutasVendedor)
```

```python
agrupaDataVendedor = vendas_DF.groupby(['Data Venda', 'Vendedor'])
display(agrupaDataVendedor)
```
